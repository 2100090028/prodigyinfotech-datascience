```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Enable inline plotting
%matplotlib inline

# Load the dataset
train = pd.read_csv('titanic_train.csv')

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(train.head())

# Check for missing values
print("\nMissing values in each column:")
print(train.isnull().sum())

# Visualize missing values using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title("Heatmap of Missing Values")
plt.show()

# Countplot of survivors
sns.set_style('whitegrid')
plt.figure(figsize=(8, 4))
sns.countplot(x='Survived', data=train)
plt.title("Countplot of Survivors")
plt.show()

# Countplot of survivors by gender
plt.figure(figsize=(8, 4))
sns.countplot(x='Survived', hue='Sex', data=train, palette='RdBu_r')
plt.title("Countplot of Survivors by Gender")
plt.show()

# Countplot of survivors by passenger class
plt.figure(figsize=(8, 4))
sns.countplot(x='Survived', hue='Pclass', data=train, palette='rainbow')
plt.title("Countplot of Survivors by Passenger Class")
plt.show()

# Distribution plot of passenger ages
plt.figure(figsize=(8, 4))
sns.histplot(train['Age'].dropna(), kde=False, color='darkred', bins=40)
plt.title("Distribution of Passenger Ages")
plt.show()

# Histogram of passenger ages
plt.figure(figsize=(8, 4))
train['Age'].hist(bins=30, color='darkred', alpha=0.3)
plt.title("Histogram of Passenger Ages")
plt.show()

# Countplot of the number of siblings/spouses aboard
plt.figure(figsize=(8, 4))
sns.countplot(x='SibSp', data=train)
plt.title("Countplot of Siblings/Spouses Aboard")
plt.show()

# Histogram of passenger fares
plt.figure(figsize=(8, 4))
train['Fare'].hist(color='green', bins=40)
plt.title("Histogram of Passenger Fares")
plt.show()

# Boxplot of age by passenger class to help with age imputation
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass', y='Age', data=train, palette='winter')
plt.title("Boxplot of Age by Passenger Class")
plt.show()

# Impute Age function
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 2:
            return 39
        elif Pclass == 4:
            return 31
        else:
            return 26
    else:
        return Age

# Apply the impute_age function
train['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis=1)

# Visualize missing values after imputation
plt.figure(figsize=(10, 6))
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title("Heatmap of Missing Values After Age Imputation")
plt.show()

# Drop 'Cabin' column
train.drop('Cabin', axis=1, inplace=True)

# Drop remaining rows with missing values
train.dropna(inplace=True)

# Display dataset information
print("\nDataset information after preprocessing:")
print(train.info())

# Convert categorical variables into dummy/indicator variables
sex = pd.get_dummies(train['Sex'], drop_first=True)
embark = pd.get_dummies(train['Embarked'], drop_first=True)
train.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
train = pd.concat([train, sex, embark], axis=1)

# Display the first few rows of the modified dataset
print("\nFirst few rows of the modified dataset:")
print(train.head())

# Split the data into training and testing sets
X = train.drop('Survived', axis=1)
y = train['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)

# Train a Logistic Regression model
logmodel = LogisticRegression(max_iter=200)
logmodel.fit(X_train, y_train)

# Make predictions
predictions = logmodel.predict(X_test)

# Evaluate the model
conf_matrix = confusion_matrix(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)
class_report = classification_report(y_test, predictions)

print("\nConfusion Matrix:")
print(conf_matrix)
print("\nAccuracy Score:", accuracy)
print("\nClassification Report:")
print(class_report)

```
